```python

import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
#print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
soup.find_all('H2')
```

```python
for i in soup.find_all('p'):
  print(i.text.strip())
```
